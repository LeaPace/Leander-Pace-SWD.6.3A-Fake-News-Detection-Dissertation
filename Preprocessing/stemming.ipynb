```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import pandas as pd
import nltk
```

```python
ppr_train = pd.read_csv("../../dataset/PreprocessedLiar/train.csv")
ppr_test = pd.read_csv("../../dataset/PreprocessedLiar/test.csv")
```

```python
ppr_train
```

```python
ps = PorterStemmer()

def Stemming(text):
    
    word_list = nltk.word_tokenize(text)
    stemmed_output = ' '.join([ps.stem(w) for w in word_list])

    return stemmed_output
```

```python
ppr_train['cleaned_statement'] = ppr_train['cleaned_statement'].apply(Stemming)
```
