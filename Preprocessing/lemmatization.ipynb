```python
import pandas as pd
import nltk
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
ppr_train = pd.read_csv("../../dataset/PreprocessedLiar/train.csv")
ppr_test = pd.read_csv("../../dataset/PreprocessedLiar/test.csv")
```

```python
ppr_train
```

Lammetizing Text with Verb

```python
def Lammetize(text):
    word_list = nltk.word_tokenize(text)
    lammetized_output = ' '.join([lemmatizer.lemmatize(w, 'v') for w in word_list])

    return lammetized_output
```

```python
ppr_train['cleaned_statement'] = ppr_train['cleaned_statement'].apply(Lammetize)
ppr_test['cleaned_statement'] = ppr_test['cleaned_statement'].apply(Lammetize)
```
