```python
import numpy as np
import pandas as pd

from sklearn import pipeline,ensemble,preprocessing,feature_extraction,metrics
```

```python
train = pd.read_csv("../dataset/PreProcessedLiar/train3.csv")
test = pd.read_csv("../dataset/PreProcessedLiar/test3.csv")
```

```python
dataset = train.append(test, ignore_index=True)
```

```python
dataset
```

```python
corpus = dataset['cleaned_statement']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vec = CountVectorizer()

# tokenize, build vocab and encode training data
word_count_cvec = count_vec.fit_transform(corpus)

word_count_cvec.shape
```

```python
#print(count_vec.get_feature_names())
```

```python
vector = count_vec.transform(corpus)
```

```python
matrix = count_vec.fit_transform(corpus)
counts = pd.DataFrame(matrix.toarray(), index=[corpus], columns=count_vec.get_feature_names())
counts
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

tf_transformer = TfidfTransformer(use_idf=True).fit(matrix)
word_count_cvec_tf = tf_transformer.transform(matrix)
word_count_cvec_tf.shape
```

```python
df1 = pd.DataFrame(word_count_cvec_tf.toarray(), columns=count_vec.get_feature_names(), index=[corpus])
df1.sample(5)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity(df1[0:1], df1)
```
